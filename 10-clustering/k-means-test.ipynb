```python
from scipy import misc
import matplotlib.pyplot as plt
import numpy as np


def show_img(img):
    # What size does the figure need to be in inches to fit the image?  
    dpi = 80    
    plt.imshow(img, cmap=plt.cm.gray)
    plt.show()
    return
```

```python
img = misc.imread("images/01-why1.jpg")
show_img(img)
```

```python
from kmeans import *

model = KMeansModel(10)
```

```python
a = (img.shape[0]*img.shape[1], 3)
reshaped_img = np.reshape(img, a)
```

```python
model.init(reshaped_img)
```

```python
model.train(reshaped_img, 50)
```

```python
clusters = model.get_clusters()
centroids = model.get_centroids()

new_img = np.zeros(img.shape)
print(img.shape)

for y in range(0, img.shape[0]):
    for x in range(0, img.shape[1]) :
        new_img[y][x] = centroids[clusters[y*len(img[1]) + x]]
show_img(new_img.astype(int))
```

```python
def test(img_path):
    img = misc.imread(img_path)
    model = KMeansModel(10)
    a = (img.shape[0]*img.shape[1], 3)
    reshaped_img = np.reshape(img, a)
    model.init(reshaped_img)
    model.train(reshaped_img, 50)
    clusters = model.get_clusters()
    centroids = model.get_centroids()

    new_img = np.zeros(img.shape)
    print(img.shape)

    for y in range(0, img.shape[0]):
        for x in range(0, img.shape[1]) :
            new_img[y][x] = centroids[clusters[y*len(img[1]) + x]]
    show_img(new_img.astype(int))

test("images/01-why3.jpg")
```
