# Testing (k)nn for regression

In this case we want to predict function values. Let us start with a linear function.

$f(x) = x$

```python
from sklearn.neighbors import RadiusNeighborsRegressor 
import numpy as np
import matplotlib.pyplot as plt

X = np.reshape( list(range(-20,-5)) + list(range(5,20)), [-1, 1] )
y = np.array(X)

T = np.reshape(list(range(-15,15)), [-1,1])
```

# Data Array

```python
np.stack((X,y), axis=-1)
```

# Compare uniform kNN to weighted kNN
- for this we consider all neighbors at all times
- Note: this is not really kNN, but rather **Radius Neighbors Regressor**
    - instead of taking the k-nearest points, we take all points in a certain radius

```python
def test(X,y,T, radius):
    for i, weights in enumerate(['uniform', 'distance']):
        knn = RadiusNeighborsRegressor(radius, weights=weights)
        y_ = knn.fit(X, y).predict(T)
    
        plt.subplot(2, 1, i + 1)
        plt.scatter(X, y, c='k', label='data')
        plt.plot(T, y_, c='g', label='prediction')
        plt.axis('tight')
        plt.legend()
        plt.title("RadiusNeighborsRegressor(radius = %i, weights = '%s')" % (radius,
                                                                weights))

    plt.tight_layout()
    plt.show()

test(X, y, T, 100)
```

## Result: kNN can perfectly approximate all points in between the given points
It struggles when not all points are observed (radius too small)

```python
test(X, y, T, 8)
```

```python
noise = np.random.normal(0,2,[30,1])
y_noise = y + noise
```

```python
test(X,y_noise, T, 100)
```

### When we observe all given datapoints, kNN is robust against noise (at least in the linear case)

kNN is extremely dependent on which data points are given
